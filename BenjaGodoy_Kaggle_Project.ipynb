## Tools Introductions

### GeoPandas Introduction
Mapping Ecuador's stores per Province in an Interactive Map

```python
stores = pd.read_csv("stores.csv")
stores.head()
```

```python
stores.rename(columns={"state":"province"}, inplace=True)
stores.head()
```

```python
url = "https://data.humdata.org/dataset/e66dbc70-17fe-4230-b9d6-855d192fc05c/resource/6fa37b41-ad28-40a6-9641-3b4efd4dbe13/download/ecuador.geojson"
ecuador = gpd.read_file(url)
ecuador
```

```python
ecuador.plot(figsize=(12, 10));
```

```python
galapagos = ecuador.cx[-92:-90, -1:1]
galapagos.plot(figsize=(6, 6));
```

```python
continental = ecuador.cx[-82:, :]
continental.plot(figsize=(8, 8));
```

```python
continental
```

```python
continental.DPA_DESPRO.unique()
```

```python
stores.province.unique()
```

```python
provinces = continental.dissolve("DPA_DESPRO").loc[:, "geometry"].reset_index()
provinces.rename(columns={"DPA_DESPRO":"province"}, inplace=True)
provinces
```

```python
store_province = stores.groupby("province").count().iloc[:, 0]
store_province.index = store_province.index.str.upper()
store_province
```

```python
provinces.loc[:, "stores"] = provinces.loc[:, "province"].apply(lambda k: store_province.loc[k] if k in store_province.index else 0)
provinces
```

```python
provinces.plot(column="stores", legend=True, figsize=(8, 8));
```

```python
provinces.explore(column="stores")
```

#### Project Example

```python
train = pd.read_csv('train.csv')
train['date'] = pd.to_datetime(train['date'])
train
```

```python
hd = pd.read_csv('holidays_events.csv')
hd['date'] = pd.to_datetime(hd['date'])
hd
```

```python
sales_by_date = train[['date', 'sales']].groupby('date').sum().reset_index()
sales_by_date
```

```python
sns.scatterplot(data=sales_by_date, x='date', y='sales');
```

```python
def f1(row):
    return (hd['date'] - row['date']).abs().min().days
sales_by_date['distace_to_hd'] = sales_by_date.apply(f1, axis=1)
sales_by_date
```

```python
hd[hd['date'] == '2017-08-11']
```

```python
days_to_hd = sales_by_date[['distace_to_hd', 'sales']].groupby('distace_to_hd').sum().reset_index()
days_to_hd
```

```python
sns.histplot(data=days_to_hd, x='sales');
```

```python
sns.scatterplot(data=days_to_hd, x='distace_to_hd', y='sales');
```

```python
sns.scatterplot(data=sales_by_date, x='distace_to_hd', y='sales');
```
